```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import rdkit
from rdkit import Chem
from rdkit.Chem import Draw

%matplotlib inline
```

```python
bb1_pactive = pd.read_csv('../data_preparation/output/bb1_pactive.csv')
bb2_pactive = pd.read_csv('../data_preparation/output/bb2_pactive.csv')
bb3_pactive = pd.read_csv('../data_preparation/output/bb3_pactive.csv')

total = pd.read_csv('../data_preparation/output/total_compounds.csv')
actives = total.loc[total['read_count'] > 0]
inactives = total.loc[total['read_count'] == 0]
```

### Visualize distribution of P(active) values for each building block position

```python
fig, axs = plt.subplots(1, 3, figsize=(15,6), dpi=150)
axs[0].hist(bb1_pactive['P(active)'], log=True)
axs[0].set_xlim([-0.05, 1.05])
axs[0].set_xlabel('BB1 P(active)')
axs[0].set_ylabel('Number of BBs')
axs[0].set_title('Distribution of P(active) values for BB1')

axs[1].hist(bb2_pactive['P(active)'], color='orange', log=True)
axs[1].set_xlim([-0.05, 1.05])
axs[1].set_xlabel('BB2 P(active)')
axs[1].set_ylabel('Number of BBs')
axs[1].set_title('Distribution of P(active) values for BB2')

axs[2].hist(bb3_pactive['P(active)'], color='green', log=True)
axs[2].set_xlim([-0.05, 1.05])
axs[2].set_xlabel('BB3 P(active)')
axs[2].set_ylabel('Number of BBs')
axs[2].set_title('Distribution of P(active) values for BB3')


plt.show()
```

### Visualize top most enriched building blocks at each position


```python
bb1_sorted = bb1_pactive.sort_values(by='P(active)', ascending=False)
bb1_top = bb1_sorted[:8]
bb1_mols = [Chem.MolFromSmiles(smi) for smi in bb1_top['stereo_SMILES']]
Draw.MolsToGridImage(bb1_mols, molsPerRow=4, legends=[f'P(active): {x:.3f}' for x in bb1_top['P(active)']])
```

```python
bb2_sorted = bb2_pactive.sort_values(by='P(active)', ascending=False)
bb2_top = bb2_sorted[:8]
bb2_mols = [Chem.MolFromSmiles(smi) for smi in bb2_top['stereo_SMILES']]
Draw.MolsToGridImage(bb2_mols, molsPerRow=4, legends=[f'P(active): {x:.3f}' for x in bb2_top['P(active)']])
```

```python
bb3_sorted = bb3_pactive.sort_values(by='P(active)', ascending=False)
bb3_top = bb3_sorted[:8]
bb3_mols = [Chem.MolFromSmiles(smi) for smi in bb3_top['stereo_SMILES']]
Draw.MolsToGridImage(bb3_mols, molsPerRow=4, legends=[f'P(active): {x:.3f}' for x in bb3_top['P(active)']])
```

### Visualize pairwise trends in building block positions

```python
# Get list of building blocks at each position with P(active) greater than 0
bb1_active = bb1_pactive.loc[bb1_pactive['P(active)'] > 0]
bb2_active = bb2_pactive.loc[bb2_pactive['P(active)'] > 0]
bb3_active = bb3_pactive.loc[bb3_pactive['P(active)'] > 0]

pos_3_bb1_unique = np.zeros(len(bb3_active))
pos_3_bb2_unique = np.zeros(len(bb3_active))
for index, bb in enumerate(bb3_active['SMILES']):
    pos_3_bb1_unique[index] = len(set(actives.loc[actives['bb3'] == bb]['bb1']))
    pos_3_bb2_unique[index] = len(set(actives.loc[actives['bb3'] == bb]['bb2']))

pos_2_bb1_unique = np.zeros(len(bb2_active))
pos_2_bb3_unique = np.zeros(len(bb2_active))
for index, bb in enumerate(bb2_active['SMILES']):
    pos_2_bb1_unique[index] = len(set(actives.loc[actives['bb2'] == bb]['bb1']))
    pos_2_bb3_unique[index] = len(set(actives.loc[actives['bb2'] == bb]['bb3']))
    
pos_1_bb2_unique = np.zeros(len(bb1_active))
pos_1_bb3_unique = np.zeros(len(bb1_active))
for index, bb in enumerate(bb1_active['SMILES']):
    pos_1_bb2_unique[index] = len(set(actives.loc[actives['bb1'] == bb]['bb2']))
    pos_1_bb3_unique[index] = len(set(actives.loc[actives['bb1'] == bb]['bb3']))
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6), dpi=150)

axs[0].scatter(bb3_active['P(active)'], pos_3_bb1_unique/len(bb1_active), label='BB1')
axs[0].scatter(bb3_active['P(active)'], pos_3_bb2_unique/len(bb2_active), label='BB2')
axs[0].set_xlabel('BB3 P(active)')
axs[0].set_ylabel('Proportion of active BBs sampled')
axs[0].set_title('Diversity of BB1 and BB2 as a function of BB3 P(active)')
axs[0].legend(loc='best')

axs[1].scatter(bb2_active['P(active)'], pos_2_bb1_unique/len(bb1_active), label='BB1')
axs[1].scatter(bb2_active['P(active)'], pos_2_bb3_unique/len(bb3_active), label='BB3')
axs[1].set_xlabel('BB2 P(active)')
axs[1].set_ylabel('Proportion of active BBs sampled')
axs[1].set_title('Diversity of BB1 and BB3 as a function of BB2 P(active)')
axs[1].legend(loc='best')

axs[2].scatter(bb1_active['P(active)'], pos_1_bb2_unique/len(bb2_active), label='BB2')
axs[2].scatter(bb1_active['P(active)'], pos_1_bb3_unique/len(bb3_active), label='BB3')
axs[2].set_xlabel('BB1 P(active)')
axs[2].set_ylabel('Proportion of active BBs sampled')
axs[2].set_title('Diversity of BB2 and BB3 as a function of BB1 P(active)')
axs[2].legend(loc='best')
plt.show()
```

For plot on the far right, it appears there are two different trends. This points to how for BB1 of similar P(active) value, they pair with a much different number of building blocks to form active compounds.

```python
# Cheat a little bit and join the functional group labels and see if that makes a difference

bb1_info = pd.read_csv('../../paper_scripts/bb1_info.csv')
bb1_complete = pd.merge(bb1_active, bb1_info, left_on='SMILES', right_on='bb1_SMILES')
bb1_complete = bb1_complete[['SMILES', 'stereo_SMILES', 'P(active)', 'bb1_cat']]

bb2_info = pd.read_csv('../../paper_scripts/bb2_info.csv')
bb2_complete = pd.merge(bb2_active, bb2_info, left_on='SMILES', right_on='bb2_SMILES')
bb2_complete = bb2_complete[['SMILES', 'stereo_SMILES', 'P(active)', 'bb2_cat']]

bb3_info = pd.read_csv('../../paper_scripts/bb3_info.csv')
bb3_complete = pd.merge(bb3_active, bb3_info, on='SMILES')
bb3_complete = bb3_complete[['SMILES', 'stereo_SMILES', 'P(active)_x', 'bb3_cat']]
bb3_complete = bb3_complete.rename(columns={'P(active)_x': 'P(active)'})
```

```python
pos_3_bb1_unique = np.zeros(len(bb3_complete))
pos_3_bb2_unique = np.zeros(len(bb3_complete))
for index, bb in enumerate(bb3_complete['SMILES']):
    pos_3_bb1_unique[index] = len(set(actives.loc[actives['bb3'] == bb]['bb1']))
    pos_3_bb2_unique[index] = len(set(actives.loc[actives['bb3'] == bb]['bb2']))

pos_2_bb1_unique = np.zeros(len(bb2_complete))
pos_2_bb3_unique = np.zeros(len(bb2_complete))
for index, bb in enumerate(bb2_complete['SMILES']):
    pos_2_bb1_unique[index] = len(set(actives.loc[actives['bb2'] == bb]['bb1']))
    pos_2_bb3_unique[index] = len(set(actives.loc[actives['bb2'] == bb]['bb3']))
    
pos_1_bb2_unique = np.zeros(len(bb1_complete))
pos_1_bb3_unique = np.zeros(len(bb1_complete))
for index, bb in enumerate(bb1_complete['SMILES']):
    pos_1_bb2_unique[index] = len(set(actives.loc[actives['bb1'] == bb]['bb2']))
    pos_1_bb3_unique[index] = len(set(actives.loc[actives['bb1'] == bb]['bb3']))
```

```python
bb1_group_A_ind = list(bb1_complete.loc[bb1_complete['bb1_cat'] == 'A'].index)
bb1_group_B_ind = list(bb1_complete.loc[bb1_complete['bb1_cat'] == 'B'].index)
bb1_group_C_ind = list(bb1_complete.loc[bb1_complete['bb1_cat'] == 'C'].index)
```

```python
plt.scatter(bb1_complete.iloc[bb1_group_B_ind]['P(active)'], pos_1_bb2_unique[bb1_group_B_ind]/len(bb2_active), label='B')
plt.scatter(bb1_complete.iloc[bb1_group_C_ind]['P(active)'], pos_1_bb2_unique[bb1_group_C_ind]/len(bb2_active), label='C')
plt.scatter(bb1_complete.iloc[bb1_group_A_ind]['P(active)'], pos_1_bb2_unique[bb1_group_A_ind]/len(bb2_active), label='A')
plt.legend(loc='lower right')
plt.xlabel('BB1 P(active)')
plt.ylabel('Proportion of active BBs sampled')
plt.title('Diversity of BB2 as a function of BB1 P(active)')
plt.show()
```

```python
plt.scatter(bb1_complete.iloc[bb1_group_B_ind]['P(active)'], pos_1_bb3_unique[bb1_group_B_ind]/len(bb3_active), label='amine')
plt.scatter(bb1_complete.iloc[bb1_group_C_ind]['P(active)'], pos_1_bb3_unique[bb1_group_C_ind]/len(bb3_active), label='both')
plt.scatter(bb1_complete.iloc[bb1_group_A_ind]['P(active)'], pos_1_bb3_unique[bb1_group_A_ind]/len(bb3_active), label='carboxylic acid')
plt.legend(loc='best')
plt.xlabel('BB1 P(active)')
plt.ylabel('Proportion of active BBs sampled')
plt.title('Diversity of BB3 as a function of BB1 P(active)')
plt.show()
```
